
Loading Data into Pandas

```python
import pandas as pd


df=pd.read_csv('pokemon_data.csv')#reading csv files
df_xlsx=pd.read_excel('pokemon_data.xlsx')#reading excel files
df_txt=pd.read_csv('pokemon_data.txt',delimiter='\t')#to read txt files but delimiter needed to seperate columns
#print(df.tail(3)) three first rows
#print(df.head(3)) three last rows 
print(df_txt.head(3))
```

Reading Data in Pandas

```python
#Reading Headers
print(df.columns)
print("\n")
#Reading each column
print(df['Name'][0:6])#specify how many info by slicing
      #print(df.Name) Also works but only with one word column
print(df[['Name','Type 1','HP']][:4])# You can add more columns
print("\n")
#Reading each row
print(df.iloc[1][:3])# the first row (you can also slicing for how many columns you want)
print("\n")


```

```python
for index,row in df.iterrows():#iterate rows 
  print(index,row['Name'])#specify the "name" column
  if index==3:
    break

print(df.loc[df['Type 1'] == "Fire"][:3])# access rows that have type 1 equals to fire(slicing for specifiying number of rows)
print("\n")
#Reading a specific location (R,C)
print(df.iloc[2,1])#second row first column
```

Sorting / Describing Data

```python
df.describe() # Gives stats on data (count, mean, std...)
```

```python
df.sort_values("Name")#Sorting values by names ascending(add ascending=False for the reverse sort)
```

```python
df.sort_values(["Type 1","HP"], ascending=[1,0])# sorting data by type and hp 
```

Making Changes to data

```python
df["Total"]= df["HP"]+df["Attack"]+df["Defense"]+df["Sp. Atk"]+df["Sp. Def"]+df["Speed"]#create columm total
# df=df.drop(columns=["Total"])#delete column of total

df["Total2"]=df.iloc[:,4:10].sum(axis=1)# better way to create total (all rows,4 to 9th colums) axis=1:horizontaly

cols=list(df.columns.values)# all colums in a list
df=df[cols[:4]+[cols[-1]]+cols[4:12]]#ordering colums
df.head(5)

```

Saving our Data(Exporting into Desired Format)

```python
df.to_csv("modified.csv",index=False)#saves dataframe(index=F means with no index column) 

#df.to_excel('modified.xlsx',index=False)

df.to_csv("modified2.txt",index=False,sep="\t")#you can seperate between columns
```

Filtering Data

```python
new_df= df.loc[(df["Type 1"]=='Grass') & (df["Type 2"]=='Poison') & (df["HP"]>70)]# rows that have grass on type 1 and Poinson on type 2 and hp superior than 70
new_df.reset_index(inplace=True)#reset index and saves the old indices(drop=True removes old indices)

new_df


```
